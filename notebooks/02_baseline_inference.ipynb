## Test 1 — BARThez vanilla

Modèle pré-entraîné sur résumé français généraliste. On l'utilise tel quel, sans aucun fine-tuning sur le sport.

```python
# Chargement du modèle (la première fois, téléchargement ~900 Mo)
barthez = BARThezSummarizer()

print("\n" + "-" * 80)
print("RÉSUMÉS BARThez VANILLA")

for i, row in samples.iterrows():
    t0 = time.time()
    summary = barthez.summarize(row["text"])
    duration = time.time() - t0

    print(f"\n[{i+1}] {row['topic']} ({duration:.1f}s)")
    print(f"  Référence : {row['summary']}")
    print(f"  BARThez   : {summary}")
```
